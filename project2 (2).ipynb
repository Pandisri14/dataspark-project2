```python
  # QUERY 2  
mycursor.execute("""SELECT 
    c.customerkey,
    COUNT(s.ordernumber) AS total_orders,
    SUM(s.quantity * p.unitpriceusd) AS total_spent,
    AVG(s.quantity * p.unitpriceusd) AS avg_order_value
FROM sales s
JOIN customers c ON s.customerkey = c.customerkey
JOIN products p ON s.productkey = p.productkey
GROUP BY c.customerkey
ORDER BY total_spent DESC  LIMIT 10""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))               
```

```python
# query 3     TOTAL SALES OVER A TIME IN YEARLY
mycursor.execute("select date_format(orderdate,'%y')as order_year, sum(quantity)as total_quantity_sold from sales group by order_year order by order_year")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))                 
```

```python
# query 4           TOP PRODUCTS BY QUANTITY SOLD
mycursor.execute("select productkey,sum(quantity)as total_quantity_sold from sales group by productkey order by total_quantity_sold DESC LIMIT 5")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))                 
```

```python
# QUERY 5          SALES BY STORE
mycursor.execute(" select storekey,sum(quantity)as total_quantity_sold from sales group by storekey order by total_quantity_sold")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))                                  
```

```python
# QUERY 6
mycursor.execute("""
    SELECT p.productname, SUM(s.quantity) AS total_quantity_sold 
    FROM sales s
    JOIN products p ON s.productkey = p.productkey 
    GROUP BY p.productname 
    ORDER BY total_quantity_sold DESC 
    LIMIT 10
""")

out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))                                                   
```

```python
#QUERY 7
mycursor.execute("""SELECT 
    st.country,
    st.state,
    SUM(s.quantity * p.unitpriceusd) AS total_sales,
    COUNT(st.storekey) AS number_of_stores
FROM 
    stores st
JOIN sales s ON st.storekey = s.storekey
JOIN products p ON s.productkey = p.productkey
GROUP BY 
    st.country, st.state
ORDER BY 
    total_sales DESC""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))       
```

```python
  #QUERY 8
mycursor.execute("""SELECT 
    p.productname,
    SUM(s.quantity) AS total_quantity_sold,
    SUM(s.quantity * (p.unitpriceusd - p.unitcostusd)) AS total_profit
FROM sales s
JOIN products p ON s.productkey = p.productkey
GROUP BY p.productname
ORDER BY total_profit DESC""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#QUERY 9
mycursor.execute("""SELECT 
    p.category,
    p.subcategory,
    SUM(s.quantity * p.unitpriceusd) AS total_sales
FROM sales s
JOIN products p ON s.productkey = p.productkey
GROUP BY p.category, p.subcategory
ORDER BY total_sales DESC""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))  
```

```python
#QUERY 10
mycursor.execute("""SELECT s.CustomerKey, 
       COUNT(s.OrderNumber) AS PurchaseFrequency, 
       AVG(p.UnitPriceUSD * s.Quantity) AS AvgOrderValue
FROM sales s
JOIN products p ON s.ProductKey = p.ProductKey
GROUP BY s.CustomerKey
order by PurchaseFrequency DESC LIMIT 10 """)
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))  
```
